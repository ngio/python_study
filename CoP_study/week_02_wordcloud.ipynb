```python
# 텍스트 데이터 읽어오기
text = open("D:\\_________________python\\jupyter_data\\notebook\\input_data\\pg158.txt", "rt", encoding="UTF8").read()
#text = open("pg158.txt", "rt", encoding="UTF8").read()
type(text)
text
```

```python
# 단어의 빈도를 자동으로 구해줌
wordcloud = WordCloud().generate(text)
```

```python
# 워드 클라우드 작성하기
plt.figure(figsize = (12, 12))
plt.imshow(wordcloud, interpolation = "bilinear")
plt.axis("off")
plt.show()
```

```python
# 글씨 크기를 조정한 후의 워드 클라우드 작성하기
wordcloud = WordCloud(max_font_size = 40).generate(text)
plt.figure(figsize = (12, 12))
plt.imshow(wordcloud, interpolation = "bilinear")
plt.axis("off")
plt.show()
```
