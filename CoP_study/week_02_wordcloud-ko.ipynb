```python
# 한글 텍스트 데이터 읽어오기 
file = open("D:\\_________________python\\jupyter_data\\notebook\\input_data\\negative_words_self.txt", "rt", encoding="utf-8-sig")
lists = file.readlines()
file.close()
lists
type(lists)

sentences_tag = []
 
for sentence in lists:
    print(sentence)
    sentence = re.sub("\n", "", sentence)
    sentence = sentence.strip()
    print(sentence)
    print(hannanum.nouns(sentence))
    sentences_tag.append(hannanum.pos(sentence))

sentences_tag
```

```python
morphs = [] 

for sentence in sentences_tag: 
    for word, tag in sentence :
        #if tag in ['Noun', 'Adjective'] :
        if tag in ['N', 'P'] :
            #sentence = hannanum.nouns(re.sub('[^가-힣a-zA-Z\s]', '', sentence ))  
            #sentence = [a for a in sentence if len(a) > 1]
            morphs.append(word)
    
type(morphs)
```
